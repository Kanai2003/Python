```python
import dash
import dash_core_components as dcc
import dash_html_components as html
import numpy as np
import pandas as pd
import plotly.graph_objs as go
from datetime import datetime as dt
from datetime import timedelta
```

```python
app = dash.Dash()
```

```python
np.random.seed(50)
x_rand =  np.random.randint(1,61,60)
y_rand = np.random.randint(1,61,60)
```

```python
app.layout = html.Div([
    html.H1(children= "Hello World !", style= {
        'textAlign': 'center',
        'color': '#FF6B6B'
    }), 
    html.Div(children="Dash - A data product development framework from plotly"),

    dcc.Graph(
        id='sample-chart',
        figure = {
            'data' : [
                {'x': [4, 6, 8], 'y': [12, 16, 18], 'type': 'bar', 'name': 'First Chart'},
                {'x': [5, 7, 9], 'y': [22, 26, 28], 'type': 'bar', 'name': 'Second Chart'}
            ],
            'layout': {
                'title' : "Simple Bar Chart"
            }
        }
    ),

    dcc.Graph(
        id='scatter_chart',
        figure = {
            'data' : [
                go.Scatter(
                    x = x_rand,
                    y = y_rand,
                    mode ='markers'
                )
            ],
            'layout' : go.Layout(
                title = 'Scatter Plot of Random 60 points',
                xaxis = {'title': 'Random X values'},
                yaxis = {'title': 'Random Y values'}, 
            )
        }
    ),

    html.Label('Choose a City'),
    dcc.Dropdown(
        id = 'first-dropdown',
        options = [
            {'label':'kolkata','value':'kol'},
            {'label':'delhi','value':'del'},
            {'label':'chennai','value':'chen'},
        ],
        placeholder = "Select a city",
        multi = True
    ),

    html.Label('This is a Slider'),
    dcc.Slider(
        min = 2,
        max = 10,
        marks = {i : i for i in range(10)},
        step = 0.5,
        value = 5
    ),
    html.Label('This is a range slider'),
    dcc.RangeSlider(
        min = 1,
        max = 10,
        step = 0.5,
        value =[3,7],
        marks = {i : i for i in range(10)}
    ),

    html.Label("Input box"),
    html.Div([
        dcc.Input(
            placeholder="Your Name",
            type = 'text',
            value = '',
        )
    ]),

    html.Label("Input textBox"),
    dcc.Textarea(
        placeholder="Your GitHub",
        # value = "GitHub",
        style={'width':'100%'}
    ),

    dcc.Checklist(
        options = [
            {'label':'kolkata','value':'kol'},
            {'label':'delhi','value':'del'},
            {'label':'chennai','value':'chen'},
        ]
    ),

    dcc.RadioItems(
        options = [
            {'label':'kolkata','value':'kol'},
            {'label':'delhi','value':'del'},
            {'label':'chennai','value':'chen'},
        ]
    ),

    html.Label("Date picker"),
    html.Div([
        dcc.DatePickerSingle(
            id='dt-pick-single',
            date = dt.today(),
        ),
    ]),

    html.Label("Date picker Range"),
    html.Div([
        dcc.DatePickerRange(
            id='dt-pick-range',
            start_date = dt.today()-timedelta(days = 1),
            end_date = dt.today(),
        ),
    ])

    html.Div([
        html.Button('Submit',id = 'submit-form'),
    ]),
    


])
```

```python
if __name__ == "__main__": 
    app.run_server(port = 4050)
```
